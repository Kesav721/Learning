```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

```python
df = pd.read_csv("IPL Player Stat.csv")
df.head()
```

```python
df.columns
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
ipl_df_cleaned = df.drop(columns=["Unnamed: 0"])
missing_values = ipl_df_cleaned.isnull().sum()
```

Top Performing Batsmen

```python
num_players=ipl_df_cleaned.shape[0]

unique_players =ipl_df_cleaned['player'].nunique()

num_players, unique_players

```

Top 10 run-scorers

```python
top_scorers = ipl_df_cleaned[['player', 'runs']].sort_values(by='runs', ascending=False)
top_scorers.head(10)

plt.figure(figsize=(10, 6))
top_batsmen = top_scorers.sort_values(by='runs', ascending=False).head(10)
sns.barplot(data=top_batsmen, x='runs', y='player', palette='Blues_d')
plt.title('Top 10 Batsmen by Total Runs')
plt.xlabel('Runs')
plt.ylabel('Player')
plt.show()
```

Top 10 batting Averages

```python
top_avg = ipl_df_cleaned[['player', 'batting_avg']].sort_values(by='batting_avg', ascending=False)
top_avg.head(10)
```

Top 10 strike rates


```python
top_strike = ipl_df_cleaned[['player', 'batting_strike_rate']].sort_values(by='batting_strike_rate', ascending=False)
top_strike.head(10)
```

Bowling Performance Analysis

```python
# Top 10 wicket-takers
top_wickets = ipl_df_cleaned[['player', 'wickets']].sort_values(by='wickets', ascending=False).head(10)

# Best bowling average
top_bowl_avg = ipl_df_cleaned[['player', 'bowling_avg']].replace(0, float('inf')).sort_values(by='bowling_avg').head(10)

# Best economy rates
top_economy = ipl_df_cleaned[['player', 'bowling_economy']].sort_values(by='bowling_economy').head(10)

top_wickets, top_bowl_avg, top_economy
```

Fielding Performance Analysis

```python
# Top catchers
top_catches = ipl_df_cleaned[['player', 'catches']].sort_values(by='catches', ascending=False).head(10)\
# Top stumping
top_stumpings = ipl_df_cleaned[['player', 'stumpings']].sort_values(by='stumpings', ascending=False).head(10)

top_catches, top_stumpings
```

Batting Average vs Strike Rate

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=ipl_df_cleaned, x='batting_strike_rate', y='batting_avg', hue='runs', size='runs', sizes=(20, 200), palette='cool')
plt.title('Batting Average vs Strike Rate')
plt.xlabel('Strike Rate')
plt.ylabel('Batting Average')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
