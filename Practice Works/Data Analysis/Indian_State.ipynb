```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Indian States.csv")
```

```python
df.head()
```

```python
df.columns = [
    "State/UT", "Income_2000", "Income_2011",
    "Literacy_2001", "Literacy_2011",
    "Population_2001", "Population_2011",
    "SexRatio_2001", "SexRatio_2011",
    "Unemployment_2001", "Unemployment_2011",
    "Poverty_2001", "Poverty_2011"
]
```

```python
fig, axs = plt.subplots(3, 2, figsize=(18, 16))
fig.suptitle("State-wise Socio-Economic Indicators (2011)", fontsize=20, fontweight='bold')
sns.barplot(y="State/UT", x="Income_2011", data=df.sort_values("Income_2011", ascending=False), ax=axs[0, 0], palette="Blues_d")
axs[0, 0].set_title("Per Capita Income (2011)")
axs[0, 0].set_xlabel("Income")
axs[0, 0].set_ylabel("State/UT")
sns.barplot(y="State/UT", x="Literacy_2011", data=df.sort_values("Literacy_2011", ascending=False), ax=axs[0, 1], palette="Greens_d")
axs[0, 1].set_title("Literacy Rate (2011)")
axs[0, 1].set_xlabel("Literacy Rate (%)")
axs[0, 1].set_ylabel("")
sns.barplot(y="State/UT", x="Population_2011", data=df.sort_values("Population_2011", ascending=False), ax=axs[1, 0], palette="Oranges_d")
axs[1, 0].set_title("Population (2011)")
axs[1, 0].set_xlabel("Population")
axs[1, 0].set_ylabel("State/UT")
sns.barplot(y="State/UT", x="SexRatio_2011", data=df.sort_values("SexRatio_2011", ascending=False), ax=axs[1, 1], palette="Purples_d")
axs[1, 1].set_title("Sex Ratio (2011)")
axs[1, 1].set_xlabel("Females per 1000 Males")
axs[1, 1].set_ylabel("")
sns.barplot(y="State/UT", x="Unemployment_2011", data=df.sort_values("Unemployment_2011", ascending=False), ax=axs[2, 0], palette="Reds_d")
axs[2, 0].set_title("Unemployment Rate (2011)")
axs[2, 0].set_xlabel("Unemployment Rate (%)")
axs[2, 0].set_ylabel("State/UT")
sns.barplot(y="State/UT", x="Poverty_2011", data=df.sort_values("Poverty_2011", ascending=False), ax=axs[2, 1], palette="BuPu_d")
axs[2, 1].set_title("Poverty Rate (2011)")
axs[2, 1].set_xlabel("Poverty Rate (%)")
axs[2, 1].set_ylabel("")
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

```python
df["Income_Growth_%"] = ((df["Income_2011"] - df["Income_2000"]) / df["Income_2000"]) * 100
df["Literacy_Growth_%"] = df["Literacy_2011"] - df["Literacy_2001"]
df["Population_Growth_%"] = ((df["Population_2011"] - df["Population_2001"]) / df["Population_2001"]) * 100
df["SexRatio_Change"] = df["SexRatio_2011"] - df["SexRatio_2001"]
df["Unemployment_Change"] = df["Unemployment_2011"] - df["Unemployment_2001"]
df["Poverty_Change"] = df["Poverty_2011"] - df["Poverty_2001"]
```

```python
df.head()
```

```python
print("\nTop 5 states by Income Growth:")
print(df.sort_values("Income_Growth_%", ascending=False)[["State/UT", "Income_Growth_%"]].head())

print("\nTop 5 states by Literacy Improvement:")
print(df.sort_values("Literacy_Growth_%", ascending=False)[["State/UT", "Literacy_Growth_%"]].head())

print("\nTop 5 states by Poverty Reduction:")
print(df.sort_values("Poverty_Change")[["State/UT", "Poverty_Change"]].head())
```

```python
plt.figure(figsize=(10, 6))
df_sorted = df.sort_values("Income_2011", ascending=False)
plt.barh(df_sorted["State/UT"], df_sorted["Income_2011"], label='2011')
plt.barh(df_sorted["State/UT"], df_sorted["Income_2000"], label='2000', alpha=0.6)
plt.xlabel("Per Capita Income")
plt.title("Income Comparison (2000 vs 2011)")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(y="State/UT", x="Literacy_Growth_%", data=df.sort_values("Literacy_Growth_%", ascending=False))
plt.title("Literacy Rate Growth (2001 to 2011)")
plt.xlabel("Growth in %")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(y="State/UT", x="Poverty_Change", data=df.sort_values("Poverty_Change"))
plt.title("Poverty Rate Change (2001 to 2011)")
plt.xlabel("Change in Poverty %")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(y="State/UT", x="Population_Growth_%", data=df.sort_values("Population_Growth_%", ascending=False))
plt.title("Population Growth % (2001 to 2011)")
plt.xlabel("Growth in Population %")
plt.tight_layout()
plt.show()
```
