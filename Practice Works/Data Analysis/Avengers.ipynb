```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('avengers.csv',encoding='ISO-8859-1')
df.head()
```

```python
df.shape
selected_columns = ['Name/Alias', 'Death1', 'Return1', 'Death2', 'Return2']
print(df[selected_columns])
```

```python
df.isnull().sum()
```

```python
df = df.drop(columns=['URL','Death2','Death3','Return1', 'Return2','Return3', 'Return4', 'Return5', 'Death4', 'Death5'])
df.shape
```

```python
df.isnull().sum()
```

```python
#Distribution of Appearance
plt.figure(figsize=(10, 6))
sns.histplot(df['Appearances'], bins=30, kde=True, color='skyblue')
plt.title("Distribution of Avengers Appearances")
plt.xlabel("Number of Appearances")
plt.ylabel("Count")
plt.show()

#Appearance by Gender
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='Gender', y='Appearances', palette='pastel')
plt.title("Appearances by Gender")
plt.show()
```

```python
gender_counts = df['Gender'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=140)
plt.title("Gender Distribution")
plt.axis('equal')
plt.show()
```

```python
#Avengers by Appearance
top10 = df.sort_values(by='Appearances', ascending=False).head(10)
plt.figure(figsize=(12, 6))
sns.barplot(data=top10, x='Name/Alias', y='Appearances', palette='magma')
plt.title("Top 10 Avengers by Appearances")
plt.xticks(rotation=45)
plt.show()
top10.head()
```
