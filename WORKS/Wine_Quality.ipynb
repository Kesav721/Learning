```python
plt.figure(figsize=(6,4))
sns.countplot(x='quality', data=df)
plt.title("Distribution of Wine Quality")
plt.xlabel("Quality")
plt.ylabel("Count")
plt.show()
```

```python
df.hist(bins=15, figsize=(15, 10), layout=(4, 4), color='teal')
plt.suptitle("Histogram of All Features")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15, 12))
for i, column in enumerate(df.columns.drop('quality')):
    plt.subplot(4, 3, i+1)
    sns.boxplot(x='quality', y=column, data=df)
    plt.title(f'{column} by Wine Quality')
plt.tight_layout()
plt.show()

```
