```python
selected_countries = ['United States', 'China', 'India', 'Japan']
df_selected = df[df['Country Name'].isin(selected_countries)].set_index('Country Name')
df_selected = df_selected.drop(columns=['Country Code'])
df_selected = df_selected.transpose()
df_selected = df_selected.apply(pd.to_numeric, errors='coerce')
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(14, 7))
for country in selected_countries:
    plt.plot(df_selected.index, df_selected[country], label=country)

plt.title('GDP Over Time (1960-2020)')
plt.xlabel('Year')
plt.ylabel('GDP (in USD)')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.grid(True)
plt.show()
```
