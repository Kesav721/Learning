```python
team_counts = df['Team'].value_counts()
team_percent = (team_counts / len(df) * 100)

print(team_counts)
print(team_percent)
```

```python
team_counts.plot(kind='bar')
plt.title('Employee Count by Team')
plt.xlabel('Team')
plt.ylabel('Number of Employees')
plt.show()
```

New Orleans Pelicans team have the maximum number of employees with 19 employees.
Orlando Magic and Minnesota Timberwolves , these two teams have the minimum number of employees with 14 employees


```python
position_counts = df['Position'].value_counts()
print(position_counts)
```

The position 'SG' have 102 employee which is the maximum.
The position 'C' have 79 employee which is the minimum.

```python
position_counts.plot(kind='bar')
plt.title('Employee Count by Position')
plt.xlabel('Position')
plt.ylabel('Number of Employees')
plt.show()
```

```python
bins = [18, 25, 30, 35, 40, 45, 50, 60, 100]
labels = ['18-24', '25-29', '30-34', '35-39', '40-44', '45-49', '50-59', '60+']
df['age_group'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)
df
```

```python
agegroup_counts = df['age_group'].value_counts()
agegroup_counts
```

The data is categorized into their respective age groups. There are 182 people in 25-29 age group. Maximum number of employees are in this age group. There are no employees with age more than 45.

```python
agegroup_counts.plot(kind='bar')
plt.title('Employee Count by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Number of Employees')
plt.show()
```

```python
team_salary = df.groupby('Team')['Salary'].sum().sort_values(ascending=False)
position_salary = df.groupby('Position')['Salary'].sum().sort_values(ascending=False)

print("Total Salary Expenditure by Team:")
print(team_salary)

print("\nTeam with Highest Salary Expenditure:")
print(team_salary.idxmax(), ":", team_salary.max())

print("\nTotal Salary Expenditure by Position:")
print(position_salary)

print("\nPosition with Highest Salary Expenditure:")
print(position_salary.idxmax(), ":", position_salary.max())
```

```python
team_salary.plot(kind='bar')
plt.title('Total Salary Expenditure by Team')
plt.xlabel('Team')
plt.ylabel('Total Salary')
plt.show()

position_salary.plot(kind='bar')
plt.title('Total Salary Expenditure by Position')
plt.xlabel('Position')
plt.ylabel('Total Salary')
plt.show()
```

The team Cleveland Cavaliers has the highest salary, which is 106988689 whereas,Position C is the position with Highest Salary Expenditure of 466377332.0
